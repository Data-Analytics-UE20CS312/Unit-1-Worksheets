# Script to modify original Charles Book Club CSV 

```python
import pandas as pd
import numpy as np
import os
```

## Load data into DataFrame 

```python
pwd!
```

```python
org_df = pd.read_csv(os.getcwd()+'/CharlesBookClub.csv')
org_df
```

## Replace catergorical with text 

### Make deep copy of DataFrame

```python
new_df =org_df.copy()
new_df
```

### Replace Gender with Text 

```python
# Replace Gender (0,1) with (Male, Female)
# Male    = 0
# Female  = 1
new_df['Gender'] = org_df['Gender'].replace([0],'Male')
new_df['Gender'] = new_df['Gender'].replace([1],'Female')
new_df['Gender']
```

### Replace MCode

```python
# Replace MCode (1,2,3,4,5) with (0-25, 26-50, 51-100, 101-200, 200+)
new_df['Mcode'] = org_df['Mcode'].replace([1],'0-25')
new_df['Mcode'] = new_df['Mcode'].replace([2],'26-50')
new_df['Mcode'] = new_df['Mcode'].replace([3],'51-100')
new_df['Mcode'] = new_df['Mcode'].replace([4],'101-200')
new_df['Mcode'] = new_df['Mcode'].replace([5],'200+')
new_df['Mcode']
```

## Replace RCode

```python
# Replace RCode (1,2,3,4) with (0-2 mos., 3-6 mos., 7-12 mos., 13+ mos.)
new_df['Rcode'] = org_df['Rcode'].replace([1],'0-2 mos.')
new_df['Rcode'] = new_df['Rcode'].replace([2],'3-6 mos.')
new_df['Rcode'] = new_df['Rcode'].replace([3],'7-12 mos.')
new_df['Rcode'] = new_df['Rcode'].replace([4],'13+ mos.')
new_df['Rcode']
```

## Replace FCode

```python
# Replace Fcode (1,2,3) with (1 book, 2 books, 3+ books)
new_df['Fcode'] = org_df['Fcode'].replace([1],'1 book')
new_df['Fcode'] = new_df['Fcode'].replace([2],'2 books')
new_df['Fcode'] = new_df['Fcode'].replace([3],'3+ books')
new_df['Fcode']
```
