# Script to modify original Charles Book Club CSV 

```python
import pandas as pd
import numpy as np

import os
import random
```

## Load data into DataFrame 

```python
pwd!
```

```python
org_df = pd.read_csv(os.getcwd()+'/cbc_dataset.csv')
org_df
```

## Replace catergorical with text 

### Make deep copy of DataFrame

```python
new_df = org_df.copy()
new_df
```

### Remove Code columns

```python
new_df = org_df.drop(['Mcode','Rcode','Fcode'], axis=1)
new_df
```

## Introduce Noise in M R F

```python
# Remove 93 M rows in random
remove_n = 93
drop_indices = np.random.choice(new_df.index, remove_n, replace=False)
new_df['M'] = new_df['M'].drop(drop_indices)
new_df
```

```python
# Remove 342 R rows in random
remove_n = 342
drop_indices = np.random.choice(new_df.index, remove_n, replace=False)
new_df['R'] = new_df['R'].drop(drop_indices)
new_df
```

```python
# Remove 218 F rows in random
remove_n = 218
drop_indices = np.random.choice(new_df.index, remove_n, replace=False)
new_df['F'] = new_df['F'].drop(drop_indices)
new_df
```

## Introduce Random Columns for Data Reduction

### Introduce Name, Email & Hobbies

```python
from faker import Faker

faker = Faker()

name_column = []
phone_column = []
job_column = []
addr_column = []

for i in range(len(new_df)):
    name_column.append(faker.name())
    addr_column.append(faker.address())
    phone_column.append(faker.phone_number())
    job_column.append(faker.job())
    
new_df['Name'] = name_column
new_df['Phone_No.'] = phone_column
new_df['Address'] = addr_column
new_df['Job'] = job_column

new_df
```

## Store new CSV 

```python
new_df.to_csv(os.getcwd()+'/CharlesBookClub_R.csv')
```
